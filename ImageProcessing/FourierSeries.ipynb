```python
import matplotlib.pyplot as plt
import numpy as np
import ipywidgets as widgets
from IPython.display import display

# Function to plot the sum of Fourier components
def plot_fourier_series(num_terms):
    # Time axis
    t = np.linspace(-np.pi, np.pi, 400)

    # Initialize the Fourier series
    fourier_series = np.zeros_like(t)

    # Sum of Fourier components
    for n in range(1, num_terms + 1):
        fourier_series += (1/n) * np.sin(n * t)

    # Plot
    plt.figure(figsize=(10, 4))
    plt.plot(t, fourier_series)
    plt.xlabel('Time')
    plt.ylabel('Amplitude')
    plt.title(f'Sum of first {num_terms} Fourier Components')
    plt.grid(True)
    plt.show()

# Widgets for interaction
num_terms_widget = widgets.IntSlider(value=1, min=1, max=100, step=1, description='Number of Terms:')

# Interactivity
widgets.interactive(plot_fourier_series, num_terms=num_terms_widget)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import ipywidgets as widgets
from IPython.display import display

# Function to plot the sinusoid
def plot_sinusoid(period, sampling_period):
    maxTime = 30
    # Time axis
    t = np.linspace(0, maxTime, 1000)  # 2 periods for better visualization
    # Sinusoid
    sinusoid = np.sin(2 * np.pi * t / period)

    # Sampling points
    sample_times = np.arange(0, maxTime, sampling_period)
    sample_values = np.sin(2 * np.pi * sample_times / period)

    # Plot
    plt.figure(figsize=(10, 4))
    plt.plot(t, sinusoid, label='Sinusoid')
    plt.scatter(sample_times, sample_values, color='red', label='Samples')
    plt.xlabel('Time')
    plt.ylabel('Amplitude')
    plt.title('Sampling of a Sinusoid')
    plt.legend()
    plt.grid(True)
    plt.show()

# Widgets for interaction
period_widget = widgets.FloatSlider(value=2, min=0.1, max=10, step=0.1, description='Period:')
sampling_period_widget = widgets.FloatSlider(value=0.2, min=0.01, max=10, step=0.1, description='Sampling Period:')

# Interactivity
widgets.interactive(plot_sinusoid, period=period_widget, sampling_period=sampling_period_widget)
```
